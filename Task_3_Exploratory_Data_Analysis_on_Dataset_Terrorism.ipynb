# **Feature Selection**

```python
data = dataset[["eventid","iyear", "imonth", "iday","gname", "country_txt", "provstate", "region_txt", 
           "longitude", "latitude", "attacktype1_txt", "target1", "nkill", "nwound", 
           "summary", "gname", "targtype1_txt", "weaptype1_txt", "motive"]]
```

**Renaming the columns of the dataset**

```python
#renaming
data.rename(columns={'eventid':'eventNo.','iyear':'Year','imonth':'Month','iday':'Day','gname':'groupname','country_txt':'Country','provstate':'State', 
                    'region_txt':'Region','longitude':'Longitude', 'latitude':'Latitude', 'attacktype1_txt':'AttackType',
                    'target1':'Target','nkill':'Killed','nwound':'Wounded','summary':'Summary','gname':'Group',
                    'targtype1_txt':'Target_type', 'weaptype1_txt':'Weapon_type','motive':'Motive'},inplace=True)
```

# **Data Visualizations:**

```python
plt.subplots(figsize=(15,6))
sns.barplot(data['Country'].value_counts()[:20].index,data['Country'].value_counts()[:20].values,palette = 'dark')
plt.title('Top 20 Countries that are Affected')
plt.xlabel('Countries')
plt.ylabel('Count')
plt.xticks(rotation= 90)
plt.show()
```

```python
x_year = data['Year'].unique()
y_count_years = data['Year'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (18,8))
sns.barplot(x = x_year, y = y_count_years, palette = 'rocket_r')
plt.xticks(rotation = 45)
plt.xlabel('Attack Year')
plt.ylabel('Number of Attacks in each year')
plt.title('Number of attacks year wise')
plt.show()
```

```python
plt.subplots(figsize=(12,6))
sns.barplot(data['Region'].value_counts().index,data['Region'].value_counts().values)
plt.title('Top Regions that are Affected')
plt.xlabel('Regions')
plt.ylabel('Count')
plt.xticks(rotation= 90)
plt.show()
```

```python
plt.subplots(figsize=(15,6))
sns.barplot(data['Country'].value_counts()[:15].index,data['Country'].value_counts()[:15].values,palette = 'viridis')
plt.title('Top Countries Affected')
plt.xlabel('Countries')
plt.ylabel('Count')
plt.xticks(rotation= 90)
plt.show()
```

```python
plt.subplots(figsize=(20,6))
sns.barplot(data['State'].value_counts()[:25].index,data['State'].value_counts()[:25].values,palette='YlOrBr_r')
plt.title('Top States that are Affected')
plt.xlabel('States')
plt.ylabel('Count')
plt.xticks(rotation= 90)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=data['AttackType'].value_counts().index,y = data['AttackType'].value_counts().values, palette = 'dark')
plt.title('common Attack Type')
plt.xlabel('Count')
plt.xticks(rotation= 90)
plt.ylabel('Attack Type')
plt.show()
```

```python
Death_each_year = data.groupby(['Year'], dropna = False)['Killed'].sum() 
fig=plt.figure(figsize=(20,7))
Death_each_year.plot(kind = 'bar', colormap = "Set2")
plt.xlabel("Years", fontsize = 12)
plt.ylabel("Number of deaths",fontsize = 12)
plt.xticks(rotation = 90)
plt.title("Number of Deaths each year")
plt.show()
```

```python
fig=plt.figure(figsize=(17,7))
sns.set_style("darkgrid")
sns.kdeplot(data['Year'], hue = data['Region'])
plt.xlabel("Years")
plt.ylabel("Number of attacks")
plt.title("Number of Terrorist Attacks by Region in each year")
```
